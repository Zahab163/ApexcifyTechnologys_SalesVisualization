## Visualize Monthly Sales

### Objective
Read company sales data and plot a line chart showing the sales trend over the year.


 Visualize monthly sales
 * Read sales data of a company Plot a line chart showing the sales trend over the year tools pandas , python ,matplotlib sample
 * columns: months,  sales
 * Key Concept : Data visualization plot(), xlabel(), ylabel(), title()

Read CSV Data




```python
import pandas as pd

```

```python
# Load sales data
sales_df = pd.read_csv("/content/Sales_Data.csv")
```

```python
# Display the Data Frame
sales_df.head()
```

## Plot  Charts

```python
# Plot Line Chart
import matplotlib.pyplot as plt
import seaborn as sns

plt.plot(sales_df["Month"], sales_df["Sales_Actual"], marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Plot objectives vs actuals
plt.figure(figsize=(10,6))
sns.lineplot(x="Month", y="Sales_Objective", data=sales_df, marker="o", label="Objective")
sns.lineplot(x="Month", y="Sales_Actual", data=sales_df, marker="s", label="Actual")

plt.title("Monthly Sales Objectives vs Actuals")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.legend()
plt.show()
```
